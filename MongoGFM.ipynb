```python
#Distribute regions for the multiple threads
def BatchGenerator(l, n):
    tid = 0
    for i in range(0, len(l), n):
        tid = tid + 1
        yield (l[i:i + n], 'Thread_'+str(tid))
        
for item in BatchGenerator(docid, 2000):
    print(len(item[0]),item[1])
```
