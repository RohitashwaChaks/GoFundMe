```python
import os

directory = "../images/"
files = []
for filename in os.listdir(directory):
    if filename.endswith(".jpg"):
        files.append(os.path.join(directory, filename))
print(len(files))
```

```python
import pandas as pd
df = pd.DataFrame(columns = ["File","FaceCount","Location","Emotions"])
df["File"] = files
df.head()
```
